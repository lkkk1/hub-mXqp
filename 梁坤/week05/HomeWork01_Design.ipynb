## 作业一：后端与算法开发角色任务总结

### 核心目标: 实现高效准确的用户提问匹配,完成FAQ客服智能问答系统的搭建

### 后端开发任务：
1. 数据库设计：需建立类目表（支持多级分类）和FAQ表。FAQ表需包含标准问、相似问集合（上限200条）、答案、生效时间、版本环境（测试/正式）等字段。需设计双环境的发布机制。
2. API接口开发：提供类目管理、FAQ增删改查、相似度匹配查询等接口。
3. 系统架构：采用微服务架构，支持高并发检索。需实现向量数据的持久化存储和增量更新机制。

### 算法开发任务：
1. 模型选型：采用Sentence-BERT作为文本编码器，该模型在语义相似度任务上表现优异且计算效率高，支持中英文混合场景。
2. 文本向量化：将FAQ库中所有问句编码为768维向量，存入向量数据库（如Milvus）。
3. 相似度计算：使用余弦相似度进行向量检索，返回Top-K最相似结果。
4. 优化策略：设置相似度阈值（如0.75），低于阈值则转人工客服；通过领域数据微调模型提升业务适配性。

协同点：后端需为算法提供数据同步接口，算法需为后端提供实时匹配服务。整套方案不使用大模型，因为检索式问答无需生成能力，BERT级模型在效果和成本间已达最优平衡。
